```python
from redcap import Project
import pandas as pd

# Define your REDCap URL and API token
REDCAP_URL = 'https://vredcap.aku.edu/redcap/api/'
API_TOKEN = 'EBBDE24FD6AB103260A018339C8FF624'  # Ensure this token has the correct permissions

# Initialize the Project
project = Project(REDCAP_URL, API_TOKEN)

# Export records with labels
data = project.export_records(format_type='csv', raw_or_label='label', raw_or_label_headers='label')

# Convert data to a pandas DataFrame
df = pd.DataFrame([x.split(',') for x in data.split('\n')])

# Assuming the first row is the header
df.columns = df.iloc[0]
df = df[1:]

# Save data to a CSV file
df.to_csv('Death-Information-form_data_with_labels.csv', index=False)

print("Data extraction complete. Data saved to Death-Information-form_data_with_labels.csv.")
```
